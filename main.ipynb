```python
import pandas as pd
import os

import numpy as np
```

```python
raw_patstat = pd.read_csv(os.path.join("data", "patstat_data.csv"))

raw_patstat.columns
```

```python
profi_patstat_translations = pd.read_excel(os.path.join("data", "profi_patstat_translation.xlsx"))

profi_patstat_translations
```

```python
raw_profi = pd.read_excel(os.path.join("data", "Profi-AuszugNov23_erweitert_V5.xlsx"))
raw_profi
```

```python
patstat_id_to_profi = {}

all_patstat_ids = []
# Dict to match patstat person_id to profi id, some patstat_ids match to 


for i, row in profi_patstat_translations.iterrows():
  
    patstat_ids = row["Patstat person_id"]
    profi_id = row["ID"]
    if patstat_ids == "No Match Found!": continue
    
    patstat_ids = patstat_ids.strip("[]")

    patstat_ids = patstat_ids.split(",")

    patstat_ids = [int(x) for x in patstat_ids]
    
    for patstat_id in patstat_ids: 
        all_patstat_ids.append(patstat_id)
        if patstat_id in patstat_id_to_profi.keys():
            patstat_id_to_profi[patstat_id].append(int(profi_id))
        else: 
            patstat_id_to_profi[patstat_id] = [int(profi_id)]
        

k = patstat_id_to_profi.keys()

#all_patstat_ids

patstat_id_to_profi
```

```python
# Mapping the profi Id to the patstat dataset

patstat = raw_patstat.copy()

patstat["Profi Id"] = patstat["person_id"].map(lambda x: patstat_id_to_profi.get(x, ))
```
